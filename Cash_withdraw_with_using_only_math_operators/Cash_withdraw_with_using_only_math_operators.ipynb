```python
#FOR LOOP

amount = int(input("Enter the amount ₹ "))
cur = [2000,500,200,100,50,20,10,5,2,1]
data = []
for i in cur:
  data.append(amount//i)
  amount = amount%i
  print(data)
```

```python
#FOR LOOP

amount = int(input("Enter the amount ₹ "))
cur = [2000,500,200,100,50,20,10,5,2,1]
data = []
for i in cur:
  data.append(amount//i)
  amount = amount%i
print(data)
```

```python
#for & while
#here, FOR
amount = int(input("Enter the amount ₹ "))
cur = [2000,500,200,100,50,20,10,5,2,1]
data = []
for i in cur:
  data.append(amount//i)
  amount = amount%i
#here, While
try:
    i=-1
    while i<len(data):
      i=i+1
      if data[i]==0:
        continue
      print(data[i],cur[i])
except :
  pass
```
